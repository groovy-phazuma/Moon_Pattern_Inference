```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def load_image(name:str=""):
    filein=f"{root}/train-semi/{name}.JPEG"
    filein_segmentation=f"{root}/train-semi-segmentation/{name}.png"
    image=np.array(Image.open(filein))
    segmentation = np.array(Image.open(filein_segmentation))
    segmentation_id=segmentation[:, :, 1] * 256 + segmentation[:, :, 0]
    return image, segmentation_id
```

```python
name="n01440764/n01440764_2803"
image, segmentation_id = load_image(name=name)
```

```python
segmentation_id.max()
```

```python
plt.imshow(image)
plt.show()
plt.imshow(segmentation_id)
plt.show()
```
